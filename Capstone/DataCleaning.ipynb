```python
import matplotlib.pyplot as plt
import pandas as pd
import xgboost as xgb
import numpy
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("playsAfter2013v12.csv")
```

```python
df['game_id'] = df['game_id'].astype(str)
```

```python
df['drive'] = df['drive'].astype(str)
```

```python
df['GameAndDrive'] = df['game_id']+df['drive']
```

```python
onlyTDs = df[df['down'] != 1]
```

```python
onlyTDsWithID = onlyTDs.drop(['game_id','posteam','drive','down','yards_gained','td_team','GameAndDrive'], axis = 1)
```

```python
dfFirst = df[df['down'] == 1]
dfFirst.head()
```

```python
dfFirstNumPlays = dfFirst.groupby(['GameAndDrive'],as_index = False)['drive'].size()
```

```python
dfPlaysPerDown = dfFirst.groupby(['GameAndDrive'],as_index = False)['yards_gained'].sum()
```

```python
dfFirstNumPlays['FirstDowns'] = dfFirstNumPlays['size']
```

```python
dfFirstNumPlays = dfFirstNumPlays.drop(['size'], axis = 1)
```

```python
#dfPlaysPerDown.head(50)
```

```python
checkFirsts = pd.merge(dfFirstNumPlays,dfPlaysPerDown,on = 'GameAndDrive',left_index=True,right_index=True)
```

```python
checkFirsts['yards_per_first_down'] = checkFirsts['yards_gained']/checkFirsts['FirstDowns']
```

```python
checkTD = pd.merge(checkFirsts,onlyTDsWithID,on = 'GameAndDrive',how='left',left_index=True)
```
